```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color='red',marker='*')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
